```python
t = np.arange(0,12,0.5)
y1=np.sin(t)
y2=np.cos(t)
plt.grid()
plt.plot(t,y1,'y--',marker='o')
plt.plot(t,y2,'r--',marker='*')
plt.xlabel('Time')
plt.ylabel('Amp')
plt.title('SinCos')
```

```python
x = np.arange(0,5,1)
y1= x+1
y2= x+2
y3= x+3
y4= x+4
plt.subplot(1,2,1)
plt.plot(x,y1,'y--',marker='o')
plt.plot(x,y2,'r--',marker='*')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X,Y')
plt.subplot(1,2,2)
plt.plot(x,y3,'g--',marker='x')
plt.plot(x,y4,'b--',marker='p')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X,Y')
```
