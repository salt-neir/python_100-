```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
text = open(r'C:\Users\user\Desktop\wordcloud\\thelittleprince.txt').read()
text
```

```python
img = r'C:\Users\user\Desktop\wordcloud\\little.png'
Image.open(img)
```

```python
lp_mask=np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(
background_color = 'white',
mask = lp_mask,
max_font_size=50,
contour_width=5,
contour_color='black').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```
